```python
import pandas as pd

orders = pd.read_csv('C:/Users/aablyazizov/Desktop/DS/Mentoring/Brazilian E-Commerce/olist_orders_dataset.csv')
payments = pd.read_csv('C:/Users/aablyazizov/Desktop/DS/Mentoring/Brazilian E-Commerce/olist_order_payments_dataset.csv')
customers = pd.read_csv('C:/Users/aablyazizov/Desktop/DS/Mentoring/Brazilian E-Commerce/olist_customers_dataset.csv')

print(orders.shape)
print(payments.shape)
print(customers.shape)
```

```python
print(orders.columns.tolist())
print(payments.columns.tolist())
print(customers.columns.tolist())
```

```python
orders.head()
```

```python
df = pd.merge(orders, payments, on='order_id', how='left')
```

```python
df_delivered = df[df['order_status'] == 'delivered']
```

```python
arppu = df_delivered['payment_value'].sum()/df_delivered['customer_id'].nunique()
```

```python
print(arppu)
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['purchase_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
```

```python
first_purchase = orders.groupby('customer_id')['purchase_month'].min().reset_index()
first_purchase.columns = ['customer_id', 'cohort_month']
```

```python
orders = pd.merge(orders, first_purchase, on='customer_id', how='left')
```

```python
orders[['customer_id', 'purchase_month', 'cohort_month']].head(10)
```

```python
orders = pd.merge(orders, customers, on = 'customer_id', how = 'left')
```

```python
orders.columns.tolist()
```

```python
first_unique_order_date = orders.groupby('customer_unique_id')['purchase_month'].min().reset_index()
```

```python
first_unique_order_date.columns = ['customer_unique_id', 'cohort_month']
```

```python
orders = pd.merge(orders, first_unique_order_date, on = 'customer_unique_id', how = 'left')
```

```python
orders[['customer_unique_id', 'purchase_month', 'cohort_month']].head(10)
```

```python
orders['cohort_index'] = 12 * (orders['purchase_month'].dt.year - orders['cohort_month'].dt.year) + (orders['purchase_month'].dt.month - orders['cohort_month'].dt.month)
```

```python
orders[['customer_unique_id', 'purchase_month', 'cohort_month', 'cohort_index']][orders['cohort_index'] > 0].head(10)
```
